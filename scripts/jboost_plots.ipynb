```python
import numpy as np
%pylab
```

```python
X = np.zeros((101,2,5))
for i in range(5):
    tmp = np.loadtxt('error%s.data' % i, delimiter=' ')
    X[:,:,i] = tmp[:,1:]
```

```python
y = X.mean(2)
yerr = X.std(2)
```

```python
errorbar(range(101), y[:,0], yerr=yerr[:,0], capsize=0, fmt='-', label='train set')
errorbar(range(101), y[:,1], yerr=yerr[:,1], capsize=0, fmt='-', label='test set')
legend()
xlabel('Iteration')
xlim([0, 100])
ylabel('Error')
```
